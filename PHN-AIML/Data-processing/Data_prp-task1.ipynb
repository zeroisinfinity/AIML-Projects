```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'user'     : [' Sahil', 'sahil  ', 'Nirjala', 'nirjala', 'MAHESH', 'Sahil', np.nan, '!!!Error!!!'],
    'age'      : ['28', '28', '  ?', '34', '150', '28', '22', 'Unknown'],
    'spend'    : ['$100.5', '100.5', '85.30', 'not_available', '50.5', '$100.5', '999999', '0.0'],
    'joined'   : ['2023-12-01', '23/12/01', '2023-12-01', 'Dec-2023', '2023-12-02', '2023-12-01', np.nan, 'yesterday']
})

display(data)
display(data.info())
```

1 . Cleaning

```python

data['age'] = pd.to_numeric(data['age'] , errors='coerce')
data['spend'] = pd.to_numeric(data['spend'] , errors='coerce')

display(data)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
data[['age','spend']] = imputer.fit_transform(data[['age','spend']])
```

```python
data['age'] = data['age'].astype('int')
data['age']
```

```python
data['user'] = data['user'].str.strip().str.title()
```

```python
for u in range(0,data.shape[0]):
    c = 2
    for i in range(u+1,data.shape[0]):
        if data.loc[u,'user'] == data.loc[i,'user']:
            data.loc[i,'user'] += str(c)
            c+=1
    c = 0

data['user']
```
