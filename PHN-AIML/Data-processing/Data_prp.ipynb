# DATA CLEANING

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({
    'names' : ['Sahil' , 'Nirjala' , 'Mahesh' , 'Sahil' , np.nan],
    'age' : [28 , np.nan , 34 , 28 , 22],
    'purchase_amt' : [100.5, 85.3, np.nan, 100.5, 50.5],
    'date_of_purchase' : ['2023/12/01', '2023/12/01', '2023/12/02', '2023/12/01', '2023/12/03']
})

display(data)
display(data.info())
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
data[['age','purchase_amt']] = imputer.fit_transform(data[['age','purchase_amt']])
```

```python
data = data.drop_duplicates()
```

```python
data.loc[:,'date_of_purchase'] = pd.to_datetime(data['date_of_purchase'])
```

```python
display(data)
data.info()
```

# DATA INTEGRATION

```python
import pandas as pd

data1 = pd.DataFrame({
    'customer_id': [1, 2, 3],
    'name': ['John', 'Jane', 'Jack'],
    'age': [28, 34, 29]
})

data2 = pd.DataFrame({
    'customer_id': [1, 2, 4],
    'purchase_amount': [100.5, 85.3, 45.0],
    'purchase_date': ['2023/12/01', '2023/12/02', '2023/12/03']
})

merged_data = pd.merge(data1, data2, on='customer_id', how='inner')
merged_data
```

# DATA TRANSFORMATION

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder

data2 = pd.DataFrame({
    'category': ['A', 'B', 'A', 'C', 'B'],
    'numeric_column': [10, 15, 10, 20, 15]
})

scaler = StandardScaler()
data2['scaled_numberic_column'] = scaler.fit_transform(data2[['numeric_column']])

encoder = OneHotEncoder(sparse_output=False)
encoded_data = pd.DataFrame(
    encoder.fit_transform(data2[['category']]),
    columns = encoder.get_feature_names_out(['category']
    )
)
data3 = pd.concat([data2, encoded_data], axis=1)
print(data3)
```

# DATA REDUCTION

```python
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest , chi2
data4 = pd.DataFrame({
    'feature1': [10, 20, 30, 40, 50],
    'feature2': [1, 2, 3, 4, 5],
    'feature3': [100, 200, 300, 400, 500],
    'target' : [0,1,0,1,0]
})

selector = SelectKBest(chi2,k=2)
selected_features = selector.fit_transform(
                    X = data4[['feature1' , 'feature2' , 'feature3']] ,
                    y = data4['target']
)

print('Selected features (SelectKBest):')
print(selected_features)
pca = PCA(n_components=2)
pca_data = pca.fit_transform(data4[['feature1' , 'feature2' , 'feature3']])
print(pca_data)
```
