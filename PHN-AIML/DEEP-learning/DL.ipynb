```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

hrs = np.array(list(range(1, 7))).reshape(-1, 1)
marks = np.array([35, 40, 50, 55, 65, 70]).reshape(-1, 1)

model = keras.Sequential([
    layers.Input(shape=(1,)),
    layers.Dense(1)
])

model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mse')
model.fit(hrs, marks, epochs=500, verbose=0)

test_data = np.array([[7], [8], [10], [11], [20]])
prediction = model.predict(test_data)

print('Predicted Marks:\n', prediction)
```
